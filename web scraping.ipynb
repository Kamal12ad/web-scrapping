```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
res = requests.get(url)
```

```python
res
```

```python
ajith =BeautifulSoup(res.text)
```

```python
ajith
```

```python
Laptop = ajith.select('._4rR01T')
```

```python
Laptop
```

```python
Price = ajith.select('._1_WHN1')
```

```python
Price
```

```python
RAM = ajith.select('.rgWa7D:nth-child(2)')
```

```python
RAM
```

```python
Laptops_name=[]
```

```python
for item in Laptop:
    Laptops_name.append(item.text)
```

```python
Laptops_name
```

```python
Price_name=[]
```

```python
for item in Price:
    Price_name.append(item.text)
```

```python
Price_name
```

```python
RAM_name=[]
```

```python
for item in RAM:
    RAM_name.append(item.text)
```

```python
RAM_name
```

```python
ac = pd.DataFrame(Laptops_name,columns=["Laptops_name"])
```

```python
ac
```

```python
ac['Price']=Price
```

```python
ac
```

```python
ac["RAM"]=RAM
```

```python
ac
```

```python
ac.to_csv("ac.csv")
```
